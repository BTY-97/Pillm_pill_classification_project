```python
import numpy as np
import imutils
import cv2

# construct the argument parser and parse the arguments
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", type=str, default="shape.png",
#     help="path to input image")
# args = vars(ap.parse_args())

# plt.figure(figsize=(15,7.5))

path = './Downloads/123/Colab Notebooks/pills_test_no_bg/'
i = '1080'
image = cv2.imread(path+i+'.jpg')
# plt.subplot(2,2,1), plt.imshow(image)
cv2.imshow("Image", image)


kernel = np.ones((5,5),np.uint8)
detection = reader.readtext(img, rotation_info=[90])[0]
a,b,c,d = detection[0]
image = image[int(a[1]):int(c[1]), int(a[0]):int(b[0])]

erosion = cv2.erode(image, kernel,iterations = 1)
erosion = cv2.fastNlMeansDenoising(erosion, None, 5, 7, 21)
erosion = cv2.GaussianBlur(erosion, (5,5), 0)
# erosion = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel, iterations=4)
# erosion = cv2.adaptiveThreshold(erosion,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,13,3)
erosion = cv2.Canny(erosion, 10, 10)
# erosion = cv2.dilate(erosion, None, iterations=2)
# erosion = cv2.morphologyEx(erosion, cv2.MORPH_CLOSE, kernel, iterations=1)
# erosion = cv2.erode(erosion, kernel, iterations = 1)

# load the image and display it


# convert the image to grayscale and threshold it
# gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# thresh = cv2.threshold(gray, 200, 255,cv2.THRESH_BINARY_INV)[1]
# plt.subplot(2,2,2), plt.imshow(thresh, cmap='gray')
cv2.imshow("erosion", erosion)

# # find the largest contour in the threshold image
cnts = cv2.findContours(erosion.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
c = max(cnts, key=cv2.contourArea)

# draw the shape of the contour on the output image, compute the
# bounding box, and display the number of points in the contour
output = image.copy()
cv2.drawContours(output, [c], -1, (0, 255, 0), 3)
(x, y, w, h) = cv2.boundingRect(c)
text = "original, num_pts={}".format(len(c))
cv2.putText(output, text, (x, y - 15), cv2.FONT_HERSHEY_SIMPLEX,0.9, (0, 255, 0), 2)

# show the original contour image
print("[INFO] {}".format(text))
# plt.subplot(2,2,3), plt.imshow(output)
cv2.imshow("Original Contour", output)
cv2.waitKey(0)

# to demonstrate the impact of contour approximation, let's loop
# over a number of epsilon sizes
for eps in np.linspace(0.001, 0.03, 10):
    # approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, eps * peri, True)
    # draw the approximated contour on the image
    output = image.copy()
    cv2.drawContours(output, [approx], -1, (0, 255, 0), 3)
    text = "eps={:.4f}, num_pts={}".format(eps, len(approx))
    cv2.putText(output, text, (x, y - 15), cv2.FONT_HERSHEY_SIMPLEX,
                0.9, (0, 255, 0), 2)
    # show the approximated contour image
    print("[INFO] {}".format(text))
    cv2.imshow("Approximated Contour", output)
    cv2.waitKey(0)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
