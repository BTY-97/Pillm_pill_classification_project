```python
pills = medicines.drop(columns=['복사품목명', '품목일련번호', '업소일련번호', '업소명',
                        '성상', '큰제품이미지', '분할선앞', '분할선뒤', '크기장축', '크기단축', '크기두께',
                        '이미지생성일자(약학정보원)', '분류번호', '분류명', '전문일반구분', '품목허가일자',
                        '표기내용앞', '표기내용뒤', '표기이미지앞', '표기이미지뒤', '변경일자', 'url'])
```

```python
medi_color_f = dict()
medi_color_b = dict()
medi_shape = dict()
medi_form = dict()
medi_text_f = dict()
medi_text_b = dict()
medi_mark_f = dict()
medi_mark_b = dict()
medi_line = dict()


for i, color in enumerate(pills.색상앞):
    if color == '-':
        continue
        
    if medi_color_f.get(color) is None:
        medi_color_f[color] = list()
    medi_color_f[color].append(pills.품목명[i])
    
for i, color in enumerate(pills.색상뒤):
    if color == '-':
        continue
        
    if medi_color_b.get(color) is None:
        medi_color_b[color] = list()
    medi_color_b[color].append(pills.품목명[i])
    
for i, shape in enumerate(pills.의약품제형):
    if shape == '-':
        continue
        
    if medi_shape.get(shape) is None:
        medi_shape[shape] = list()
    medi_shape[shape].append(pills.품목명[i])
    
for i, form in enumerate(pills.제형코드명):
    if form == '-':
        continue
    
    form = form.split('|')
    if '연질캡슐제' in form:
        form = '연질캡슐제'
    elif '경질캡슐제' in form:
        form = '경질캡슐제'
    else:
        continue
        
    if medi_form.get(form) is None:
        medi_form[form] = list()
    medi_form[form].append(pills.품목명[i])
    
medi_line['십자분할선'] = list()
medi_line['분할선'] = list()

for i, text in enumerate(pills.표시앞):
    if text == '-':
        text = '텍스트없음'
        
    
    if '십자분할선' in text:
        medi_line['십자분할선'].append(pills.품목명[i])
    elif '분할선' in text:
        medi_line['분할선'].append(pills.품목명[i])
    
    if '마크' in text:
        if medi_mark_f.get(pills.표기코드앞.iloc[i]) is None:
            medi_mark_f[pills.표기코드앞.iloc[i]] = list()
        medi_mark_f[pills.표기코드앞.iloc[i]].append(pills.품목명[i])
        
    if medi_text_f.get(text) is None:
        medi_text_f[text] = list()
    medi_text_f[text].append(pills.품목명[i])
    
for i, text in enumerate(pills.표시뒤):
    if text == '-':
        text = '텍스트없음'
        
    
    if '십자분할선' in text:
        medi_line['십자분할선'].append(pills.품목명[i])
    elif '분할선' in text:
        medi_line['분할선'].append(pills.품목명[i])
    
    if '마크' in text:
        if medi_mark_b.get(pills.표기코드뒤.iloc[i]) is None:
            medi_mark_b[pills.표기코드뒤.iloc[i]] = list()
        medi_mark_b[pills.표기코드뒤.iloc[i]].append(pills.품목명[i])
        
    if medi_text_b.get(text) is None:
        medi_text_b[text] = list()
    medi_text_b[text].append(pills.품목명[i])
```

```python
def predict(colors, shape, form, line=-1, mark=[], texts=[]):
    predict_pills = list()
    for color in colors:
        if color == '-':
            continue
        if color in medi_color_f.keys():
            predict_pills.extend(medi_color_f[color])
        if color in medi_color_b.keys():
            predict_pills.extend(medi_color_b[color])
        set(predict_pills)
    
    if shape in medi_shape.keys():
        temp = list()
        for pill in medi_shape[shape]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
    
    if '연질캡슐제' in form:
        form = '연질캡슐제'
    elif '경질캡슐제' in form:
        form = '경질캡슐제'
        
    if form in medi_form.keys():
        temp = list()
        for pill in medi_form[form]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
    else:
        temp = list()
        for pill in predict_pills:
            if pill in medi_form['연질캡슐제']:
                continue
            elif pill in medi_form['경질캡슐제']:
                continue
            temp.append(pill)
        predict_pills = temp
        
    if line != -1:
        if line in medi_line.keys():
            temp = list()
            for pill in medi_line[line]:
                if pill in predict_pills:
                    temp.append(pill)
            predict_pills = temp
        else:
            temp = list()
            for pill in predict_pills:
                if pill in medi_line['분할선']:
                    continue
                elif pill in medi_line['십자분할선']:
                    continue
                temp.append(pill)
            predict_pills = temp

    if mark[0] in medi_mark_f.keys() and mark[0] != '-':
        temp = list()
        for pill in medi_mark_f[mark[0]]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp

    if mark[1] in medi_mark_b.keys() and mark[1] != '-':
        temp = list()
        for pill in medi_mark_b[mark[1]]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
            
    
    if texts == ['-','-']:
        temp = list()
        for pill in medi_text_f['텍스트없음']:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
        temp = list()
        for pill in medi_text_b['텍스트없음']:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
            
    for text in texts:
        if text == '-':
            continue
        temp = list()
        if text in medi_text_f.keys():
            for pill in medi_text_f[text]:
                if pill in predict_pills:
                    temp.append(pill)
            
        if text in medi_text_b.keys():
            for pill in medi_text_b[text]:
                if pill in predict_pills:
                    temp.append(pill)
        predict_pills = temp
            
        if '마크' in text:
            for mark_num in pills[pills.표시앞==text].표기코드앞:
                temp2 = list()
                for pill in medi_mark_f[mark_num]:
                    temp2.append(pill)
                temp3 = list()
                for pill in temp:
                    if pill in predict_pills:
                        temp3.append(pill)
                predict_pills = temp3

            for mark_num in pills[pills.표시뒤==text].표기코드뒤:
                temp2 = list()
                for pill in medi_mark_b[mark_num]:
                    temp.append(pill)
                temp3 = list()
                for pill in temp:
                    if pill in predict_pills:
                        temp3.append(pill)
                predict_pills = temp3
                
            
            
        
    return set(predict_pills)
```

```python
score = 0
for k,v in prediction.items():
    if k in v:
        score += 1
        print(k,'Correct', len(v))
    else:
        print(k,'Fail', len(v))
print(score, '/', len(prediction))
```
