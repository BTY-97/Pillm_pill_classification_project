```python
pills = medicines.drop(columns=['복사품목명', '품목일련번호', '업소일련번호', '업소명',
                        '성상', '큰제품이미지', '분할선앞', '분할선뒤', '크기장축', '크기단축', '크기두께',
                        '이미지생성일자(약학정보원)', '분류번호', '분류명', '전문일반구분', '품목허가일자',
                        '표기내용앞', '표기내용뒤', '표기이미지앞', '표기이미지뒤', '변경일자', 'url'])
```

```python
medi_color = dict()
medi_shape = dict()
medi_form = dict()
medi_text = dict()
medi_mark = dict()
medi_line = dict()


for i, color in enumerate(pills.색상앞):
    if color == '-':
        continue
        
    if medi_color.get(color) is None:
        medi_color[color] = list()
    medi_color[color].append(pills.품목명[i])
    
for i, color in enumerate(pills.색상뒤):
    if color == '-':
        continue
        
    if medi_color.get(color) is None:
        medi_color[color] = list()
    medi_color[color].append(pills.품목명[i])
    
for i, shape in enumerate(pills.의약품제형):
    if shape == '-':
        continue
        
    if medi_shape.get(shape) is None:
        medi_shape[shape] = list()
    medi_shape[shape].append(pills.품목명[i])
    
for i, form in enumerate(pills.제형코드명):
    if form == '-':
        continue
    
    form = form.split('|')
    if '연질캡슐제' in form:
        form = '연질캡슐제'
    elif '경질캡슐제' in form:
        form = '경질캡슐제'
    else:
        continue
        
    if medi_form.get(form) is None:
        medi_form[form] = list()
    medi_form[form].append(pills.품목명[i])
    
medi_line['십자분할선'] = list()
medi_line['분할선'] = list()
medi_mark[True] = list()
medi_mark[False] = list()

for i, text in enumerate(pills.표시앞):
    if text == '-':
        continue
        
    
    if '십자분할선' in text:
        medi_line['십자분할선'].append(pills.품목명[i])
    elif '분할선' in text:
        medi_line['분할선'].append(pills.품목명[i])
    
    if '마크' in text:
        medi_mark[True].append(pills.품목명[i])
        
    if medi_text.get(text) is None:
        medi_text[text] = list()
    medi_text[text].append(pills.품목명[i])
    
for i, text in enumerate(pills.표시뒤):
    if text == '-':
        continue
        
    
    if '십자분할선' in text:
        medi_line['십자분할선'].append(pills.품목명[i])
    elif '분할선' in text:
        medi_line['분할선'].append(pills.품목명[i])
    
    if '마크' in text:
        medi_mark[True].append(pills.품목명[i])
        
    if medi_text.get(text) is None:
        medi_text[text] = list()
    medi_text[text].append(pills.품목명[i])
```

```python
def predict(color, shape, form, line, mark, text=None):
    predict_pills = list()
    if color in medi_color.keys():
        predict_pills.extend(medi_color[color])
    
    if shape in medi_shape.keys():
        temp = list()
        for pill in medi_shape[shape]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
        
    if form in medi_form.keys():
        temp = list()
        for pill in medi_form[form]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
    else:
        temp = list()
        for pill in predict_pills:
            if pill in medi_form['연질캡슐제']:
                continue
            elif pill in medi_form['경질캡슐제']:
                continue
            temp.append(pill)
        predict_pills = temp
        
    if line in medi_line.keys():
        temp = list()
        for pill in medi_line[line]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
    else:
        temp = list()
        for pill in predict_pills:
            if pill in medi_line['분할선']:
                continue
            elif pill in medi_line['십자분할선']:
                continue
            temp.append(pill)
        predict_pills = temp
        
    if mark in medi_mark.keys():
        temp = list()
        for pill in medi_mark[mark]:
            if pill in predict_pills:
                temp.append(pill)
        predict_pills = temp
    else:
        temp = list()
        for pill in predict_pills:
            if pill in medi_mark[True]:
                continue
            temp.append(pill)
        predict_pills = temp
        
    return predict_pills
```
